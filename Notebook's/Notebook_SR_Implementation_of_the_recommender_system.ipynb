### Recommendation System - Implementation

```python
# Recommendation System
import ast
from sklearn.feature_extraction.text import CountVectorizer

# Metrics
from sklearn.metrics.pairwise import cosine_similarity

# Deploy the Recommendation System
import pickle
```

```python
df.describe().round(2)
```

```python
C = df["vote_average"].mean()
print(C)
```

```python
m = df["vote_count"].quantile(0.9)
print(m)
```

```python
q_movies = df.copy().loc[df["vote_count"] >= m]
q_movies.shape
```

```python
df.shape
```

```python
def wr(dataframe, m=m, C=C):
  v = dataframe["vote_count"]
  R = dataframe["vote_average"]
  return (v/(v+m) * R) + (m/(v+m) * C)
```

```python
q_movies["score"] = q_movies.apply(wr, axis=1)
q_movies.head(5)
```

```python
q_movies[["title", "vote_count", "vote_average", "score"]].head()
```

```python
q_movies = q_movies.sort_values("score", ascending=False)
q_movies.head(5)
```

#### Transformation function

```python
def transform(t):
  t_list = []
  for i in ast.literal_eval(t):
    t_list.append(i['name'])
  return t_list
```

```python
df_movies.head(1)
```

```python
df_movies['genres'] = df_movies['genres'].apply(transform)
```

```python
df_movies.head()
```

```python
df_movies['keywords'] = df_movies['keywords'].apply(transform)
```

```python
df_movies.head()
```

```python
df_movies['cast'] = df_movies['cast'].apply(transform)
```

```python
df_movies.head()
```

```python
def find_director(t):
  t_list = []
  for i in ast.literal_eval(t):
    if i['job'] == "Director":
      t_list.append(i['name'])
  return t_list
```

```python
df_movies['crew'] = df_movies['crew'].apply(find_director)
```

```python
df_movies.head()
```

```python
def replace_empty(l):
  t_replace_list = []
  for i in l:
    t_replace_list.append(i.replace(" ", ""))
  return t_replace_list
```

```python
df_movies['cast'] = df_movies['cast'].apply(replace_empty)
```

```python
df_movies.head()
```

```python
df_movies['crew'] = df_movies['crew'].apply(replace_empty)
```

```python
df_movies.head()
```

```python
df_movies['genres'] = df_movies['genres'].apply(replace_empty)
```

```python
df_movies.head()
```

```python
df_movies['keywords'] = df_movies['keywords'].apply(replace_empty)
```

```python
df_movies.head()
```

```python
df_movies['overview'] = df['overview'].apply(lambda i:i.split())
```

```python
df_movies.head()
```

```python
df_movies['tags'] = df_movies['overview'] + df_movies['genres'] + df_movies['keywords'] + df_movies['cast'] + df_movies['crew']
```

```python
df_movies
```

```python
df3 = df_movies.drop(columns=['overview', 'genres', 'keywords', 'cast', 'crew'])
```

```python
df3.head()
```

```python
df3['tags'] = df3['tags'].apply(lambda i: " ".join(i))
```

```python
df3.head()
```

```python
vectorizer = CountVectorizer(max_features=5000, stop_words='english')
```

```python
X = vectorizer.fit_transform(df3['tags']).toarray()
X.shape
```

```python
similiarity = cosine_similarity(X)
similiarity
```

```python
df3.head()
```

```python
df3[df3['title'] == 'Spectre'].index[0]
```

```python
def recommendation_system(movie):
  index = df3[df3['title'] == movie].index[0]
  distance = sorted(list(enumerate(similiarity[index])), reverse=True, key=lambda x:x[1])
  for i in distance[1:6]:
    print(df3.iloc[i[0]].title)
```

```python
recommendation_system('Just Mercy')
```
